```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample population data
data = {
    'Country': ['USA', 'China', 'India', 'Brazil', 'Russia'],
    'Population': [331002651, 1439323776, 1380004385, 212559417, 145934462],
    'Area_sqkm': [9525067, 9596961, 3287590, 8515767, 17098242]
}

df = pd.DataFrame(data)

# Calculate population density
df['Population_Density'] = df['Population'] / df['Area_sqkm']

# Create a bar chart for population density
plt.figure(figsize=(10, 6))
plt.bar(df['Country'], df['Population_Density'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population Density (people per sq. km)')
plt.title('Population Density by Country')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Display the plot
plt.show()
```
